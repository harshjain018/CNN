# **1. Introduction**

This project explores the use of a Convolutional Neural Network (CNN) for the task of digit classification. Building upon the well-known MNIST dataset, the goal is to create a model capable of accurately identifying both handwritten digits (from a self-collected sample) and typed digits rendered in various fonts (sourced from MS Word). The project will assess the trained model's effectiveness and adaptability by evaluating its performance on these diverse image sources, thereby highlighting its potential for real-world applications.

## **2. Project Overview**


Digit recognition plays a vital role in automating numerous processes across various sectors, including finance, postal services, education, and digital document handling. Conventional character recognition systems often face challenges when dealing with variations in font styles and individual handwriting.

This project seeks to address these limitations by:

Training a Convolutional Neural Network (CNN) model using the MNIST dataset as a foundation.
Assessing the trained model's accuracy and generalization capabilities on images of typed digits (featuring a range of fonts) and self-collected handwritten digits.
Refining the CNN model with the goal of enhancing its suitability for practical digit classification tasks in real-world scenarios.

## **3. Project Implementation**

**Testing Data:**

### **3.3 CNN Model Architecture**

#### **Input Layer:**
- Accepts grayscale images with dimensions of 28x28 pixels.

#### **Convolutional Layer Blocks:**
- **First Block:**
    - A 2D Convolutional layer (`Conv2D`) with 32 filters, a 3x3 kernel size, and ReLU (Rectified Linear Unit) activation.
    - A 2D Max Pooling layer (`MaxPooling`) with a pool size of 2x2.
- **Second Block:**
    - A 2D Convolutional layer (`Conv2D`) with 64 filters, a 3x3 kernel size, and ReLU activation.
    - A 2D Max Pooling layer (`MaxPooling`) with a pool size of 2x2.
- **Third Block:**
    - A 2D Convolutional layer (`Conv2D`) with 128 filters, a 3x3 kernel size, and ReLU activation.

#### **Fully Connected Layer Section:**
- A **Flatten Layer** to transform the 3D output of the convolutional layers into a 1D vector.
- A **Dense Layer** with 128 neurons and ReLU activation.
- An output **Dense Layer** with 10 neurons (corresponding to the 10 digit classes) and Softmax activation for probability distribution over the classes.

### **3.4 Model Training & Optimization**

- **Optimization Algorithm:** Adam
- **Loss Function:** Sparse Categorical Crossentropy (suitable for integer labels)
- **Number of Training Epochs:** 5
- **Training Batch Size:** 32 (number of samples processed before updating model weights)

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocess data: Reshape images to 28x28x1 and normalize pixel values (0-1)
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# Define CNN Model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),  # First Conv Layer
    layers.MaxPooling2D((2,2)),  # First MaxPooling Layer
    layers.Conv2D(64, (3,3), activation='relu'),  # Second Conv Layer
    layers.MaxPooling2D((2,2)),  # Second MaxPooling Layer
    layers.Conv2D(128, (3,3), activation='relu'),  # Third Conv Layer
    layers.Flatten(),  # Flatten output for Dense layers
    layers.Dense(128, activation='relu'),  # Fully connected layer
    layers.Dense(10, activation='softmax')  # Output layer for 10 classes (digits 0-9)
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
print("\n Training the CNN model on MNIST dataset...\n")
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Evaluate the model on test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\n Model Evaluation Complete!\nTest Accuracy: {test_acc:.4%}")

# Predict on a sample test image
sample_img = x_test[0].reshape(1, 28, 28, 1)  # Take first test image
predicted_label = np.argmax(model.predict(sample_img))

# Display the sample image with prediction
plt.figure(figsize=(4, 4))
plt.imshow(x_test[0].reshape(28,28), cmap='gray')
plt.title(f"Predicted Label: {predicted_label}", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

# Function to display images
def display_images(image_list, folder, num_images=5):
    plt.figure(figsize=(10, 5))

    for i, img_name in enumerate(image_list[:num_images]):
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for correct color display

        plt.subplot(1, num_images, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(img_name)
        plt.axis("off")

    plt.show()

# Display images
display_images(image_files, extract_folder)
```

```python
import numpy as np
import pandas as pd
import cv2

# Function to preprocess the image for CNN model
def preprocess_image(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read as grayscale
    img = cv2.bitwise_not(img)  # Invert colors (black background, white digit)
    img_resized = cv2.resize(img, (28, 28))  # Resize to 28x28 pixels
    img_normalized = img_resized.astype('float32') / 255.0  # Normalize (0-1)

    # Reshape to match CNN input format (batch size, height, width, channels)
    img_final = np.expand_dims(img_normalized, axis=0)  # Add batch dimension
    img_final = np.expand_dims(img_final, axis=-1)  # Add channel dimension

    return img_final, img_resized  # Return processed image & resized for display

# List to store results
predictions_list = []

# Loop through all extracted images
for img_name in image_files:
    img_path = os.path.join(extract_folder, img_name)

    # Extract actual digit from the filename (last character before extension)
    actual_digit = int(img_name[-5])  # Example: "image_3.png" → digit = 3

    # Preprocess image
    img_input, img_display = preprocess_image(img_path)

    # Make prediction using CNN model
    predicted_digit = np.argmax(model.predict(img_input))

    # Store results
    predictions_list.append([img_name, actual_digit, predicted_digit])

# Convert to DataFrame
df_results = pd.DataFrame(predictions_list, columns=["Image Name", "Actual Digit", "Predicted Digit"])

# Display table
df_results
```
